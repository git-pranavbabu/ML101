```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv("house_rent.csv")
```

```python
df
```

```python
df.isnull()
```

```python
df["age"] = df["age"].fillna(df["age"].mean())
```

```python
df
```

```python
x = df[["size_sqft","bedrooms","age"]]
y = df[["rent"]]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
print(f"mean squared error : {mse}\nmodel accuracy score/r2 score : {r2}")
```

```python
plt.scatter(y_test,y_pred,color = "blue")
plt.xlabel("actual value")
plt.ylabel("predicted value")
plt.title("actual vs predicted")
plt.show()
```

in this we made a model (linear regression) to predict the 
rent base on features like size, bedrooms, and age

we can infer from the metrics that we are getting a 98% accurate results from the model
(r2 score and mean squared error is used for evaluation)
