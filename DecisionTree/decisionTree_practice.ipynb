```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

from io import StringIO
```

```python
data = """Age,Salary,Experience,ExistingLoan,CreditScore,LoanApproved
25,30000,1,Yes,600,0
28,42000,3,No,650,1
35,60000,7,No,720,1
40,80000,10,No,780,1
30,45000,5,Yes,640,0
45,90000,15,No,800,1
26,32000,2,Yes,610,0
38,70000,9,No,750,1
29,48000,4,No,670,1
34,52000,6,Yes,660,0
41,85000,12,No,790,1
27,35000,2,Yes,620,0
36,68000,8,No,740,1
31,50000,5,No,690,1
39,75000,11,No,770,1"""
```

```python
#data to df
df = pd.read_csv(StringIO(data))
df.info()
```

```python
df["ExistingLoan"] = df["ExistingLoan"].map({"Yes": 1, "No": 0})
df.info()
```

```python
le = LabelEncoder()
df["ExistingLoan"] = le.fit_transform(df["ExistingLoan"])
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop("LoanApproved", axis=1), df["LoanApproved"], test_size=0.3, random_state=42)
```

```python
#using model
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
