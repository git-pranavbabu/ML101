```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, root_mean_squared_error
```

```python
df = pd.read_csv("battery_life.csv")
```

```python
df["network"] = df["network"].map({"WiFi":0,"MobileData":1})
```

```python
df
```

```python
x = df[["screen_time","apps_used","network"]]
y = df[["battery_hours"]]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
rmse = root_mean_squared_error(y_test,y_pred)
print(f"mean squared error : {mse}\nmodel accuracy score/r2 score : {r2}\nroot mean squared error : {rmse}")
```

```python
plt.scatter(y_test,y_pred,color = "blue")
plt.xlabel("actual value")
plt.ylabel("predicted value")
plt.title("actual vs predicted")
plt.show()
```
